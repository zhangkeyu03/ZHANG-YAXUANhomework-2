# Задание 2.13 (4 балла)

Разверните массив `A` в одномерный массив. Создайте три новых массива, подсчитав скользящие средние по окнам $3$, $5$, $10$. Найдите сумму этих массивов, выровнив их размерности нулями справа

*Tip: реализуйте функцию, которая считает скользящее среднее. Используйте [np.cumsum()](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html) или [np.convolve()](https://numpy.org/doc/stable/reference/generated/numpy.convolve.html)*

```python
import numpy as np

# Генерация массива A размерности 200x50
A = np.random.randint(0, 100, size=(200, 50))

# Развернуть массив A в одномерный
A_flat = A.ravel()

# Функция для подсчета скользящего среднего с помощью np.convolve
def moving_average(arr, window_size):
    return np.convolve(arr, np.ones(window_size), mode='valid') / window_size

# Рассчитываем скользящие средние для окон 3, 5, 10
ma_3 = moving_average(A_flat, 3)
ma_5 = moving_average(A_flat, 5)
ma_10 = moving_average(A_flat, 10)

# Выровнять размерности с добавлением нулей справа
max_length = max(len(ma_3), len(ma_5), len(ma_10))

# Добавляем нули справа для выравнивания
ma_3 = np.pad(ma_3, (0, max_length - len(ma_3)), 'constant')
ma_5 = np.pad(ma_5, (0, max_length - len(ma_5)), 'constant')
ma_10 = np.pad(ma_10, (0, max_length - len(ma_10)), 'constant')

# Находим сумму этих массивов
sum_ma = ma_3 + ma_5 + ma_10

# Печать результатов
print("Скользящее среднее для окна 3:", ma_3)
print("Скользящее среднее для окна 5:", ma_5)
print("Скользящее среднее для окна 10:", ma_10)
print("Сумма скользящих средних:", sum_ma)
```

# Задание 2.14 (5 баллов)

Постройте график функции $$f(x) = (5  \sin(2x) - \cos^2(x) - 1 + |x^3| - x^2) / (x^2 + 1)$$  

и производной от этой функции.  Найдите минимум (например, c помощью `scipy`) и обозначьте его на графике. На этой же оси постройте графики скользящего среднего этой функции с окном $1$, $2$, $5$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.ndimage import uniform_filter1d

# Определяем функцию f(x)
def f(x):
    return (5 * np.sin(2 * x) - np.cos(x)**2 - 1 + np.abs(x**3) - x**2) / (x**2 + 1)

# Вычисление производной функции
def df(x):
    epsilon = 1e-5  # Маленький шаг для численного дифференцирования
    return (f(x + epsilon) - f(x - epsilon)) / (2 * epsilon)

# Генерация значений x и y для функции и производной
x = np.linspace(-10, 10, 400)
y = f(x)
dy = df(x)

# Находим минимум функции с помощью scipy.optimize.minimize
res = minimize(lambda x: f(x), 0)  # Минимизация функции начиная с точки 0
minimum_x = res.x[0]
minimum_y = f(minimum_x)

# Строим графики
plt.figure(figsize=(10, 6))

# График функции f(x)
plt.plot(x, y, label=r'$f(x)$', color='blue')

# График производной df(x)
plt.plot(x, dy, label=r"$f'(x)$", color='red')

# График скользящего среднего с окнами 1, 2, 5
sm_1 = uniform_filter1d(y, size=1)
sm_2 = uniform_filter1d(y, size=2)
sm_5 = uniform_filter1d(y, size=5)

plt.plot(x, sm_1, label=r'Skipping average (window=1)', linestyle='--', color='green')
plt.plot(x, sm_2, label=r'Skipping average (window=2)', linestyle='--', color='orange')
plt.plot(x, sm_5, label=r'Skipping average (window=5)', linestyle='--', color='purple')

# Отметим минимум на графике
plt.scatter(minimum_x, minimum_y, color='black', label=f'Minimum at x={minimum_x:.2f}, f(x)={minimum_y:.2f}')

# Настройка графика
plt.title(r'Graph of $f(x)$ and its derivative, with sliding averages')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Показываем график
plt.show()
```

# Задание 2.15 (6 баллов)

Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
from typing import List 

def sample(x: List[int], c: int) -> List[int]:
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for _ in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
            
        res.append(idx)
    return res

            
sample([50, 3, 1, 7, 20], 5)   
```

Реализуйте данный алгоритм более оптимальным способом на массивах `numpy`, без использования циклов. Проверьте корректность и идентичность алгоритмов семплирования с помощью статистических тестов.

```python
import numpy as np
from typing import List

def sample(x: List[int], c: int) -> List[int]:
    assert len(x) > 0
    
    x = np.array(x)  # Конвертируем список в numpy массив
    s = np.sum(x)  # Сумма всех элементов
    cum_sum = np.cumsum(x)  # Кумулятивная сумма
    
    # Генерируем c случайных чисел, пропорциональных сумме элементов x
    vals = np.random.random(c) * s
    
    # Находим индексы с помощью np.searchsorted, которые позволяют искать в кумулятивной сумме
    res = np.searchsorted(cum_sum, vals)
    
    return res.tolist()

# Пример использования
result = sample([50, 3, 1, 7, 20], 5)
print(result)
```

Есть следующий алгоритм "случайного блуждания"

```python
import random 

def walk(n: int):
    cur_position = 0
    for i in range(n):
        cur_position += random.choice([-3, -1, 0, 2, 3])

    return cur_position

walk(10_000)
```

# Задание 2.17 (3 балла)

Имеется функция, которая проделывает достаточно странные манипуляции с массивом `numpy`

```python
import numpy.typing as npt

def strange_conv(a: npt.NDArray[np.float64]) -> np.float64:
    x_len = a.shape[0]
    y_len = a.shape[1]

    t = 1.
    for y in range(y_len - 5):
        for x in range(x_len - 3):
            c = 1.5 * a[y+1, x+2] - a[y+5, x+3] * a[y, x] + 0.2 * a[y+4, x]
            t = 0.2 * t + 0.8 * c

    return t


a = np.random.random((10000, 10000))
strange_conv(a)
```

```python
import numpy as np
import numpy.typing as npt

def strange_conv_optimized(a: npt.NDArray[np.float64]) -> np.float64:
    x_len, y_len = a.shape
    
    a1 = a[0:y_len-5, 0:x_len-3]
    a2 = a[0:y_len-5, 0:x_len-3]  # Changed from (1:x_len-2)
    a3 = a[0:y_len-5, 0:x_len-3]
    a4 = a[4:y_len-1, 0:x_len-3]


    c = 1.5 * a1 - a2 * a3 + 0.2 * a4
    t = np.mean(c)

    return t

# Пример
a = np.random.random((10000, 10000))
result = strange_conv_optimized(a)
print(result)
```
